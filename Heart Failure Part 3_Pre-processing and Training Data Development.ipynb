```python
#dealing with the features that still have outliers
df['Creatinine_phosphokinase'] = np.log(df['Creatinine_phosphokinase'])
df['Platelets'] = np.log(df['Platelets'])
df['Serum_creatinine'] = np.log(df['Serum_creatinine'])
```

```python
#code adapted from https://medium.com/analytics-vidhya/how-to-remove-outliers-for-machine-learning-24620c4657e8#:~:text=Outliers%20badly%20affect%20mean%20and,may%20statistically%20give%20erroneous%20results.&text=Most%20machine%20learning%20algorithms%20do,in%20the%20presence%20of%20outlier.
def drop_outliers(df, feature):
    iqr = 1.5 * (np.percentile(df[feature],75) - np.percentile(df[feature], 25))
    df.drop(df[df[feature] > (iqr + np.percentile(df[feature], 75))].index, inplace =True)
    df.drop(df[df[feature] < (np.percentile(df[feature], 25) - iqr)].index, inplace =True)

drop_outliers(df, 'Platelets')
drop_outliers(df, 'Serum_creatinine')
drop_outliers(df, 'Serum_sodium')
```

```python
#scale the data so that we may use certain ML Models
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
y = df['Death_event']
x = df.drop('Death_event', axis=1)
df_ss = ss.fit_transform(x)
df_ss = pd.DataFrame(data=df_ss, columns=df.columns[:-1])
df_ss = pd.concat([df_ss,y],axis=1)
```

```python
fig, ax = plt.subplots(figsize = (15, 10))
ax = sns.boxplot(data = df_ss, width = 0.5)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
plt.show()
#some features still have outliers
```
